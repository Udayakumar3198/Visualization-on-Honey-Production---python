## Domain: Food and agriculture

### Context:
In 2006, a global concern was raised over the rapid decline in the honeybee population, an integral component
to American honey agriculture. Large numbers of hives were lost to “Colony-Collapse-Disorder”, a
phenomenon of disappearing “worker-bees” causing the remaining “hive-colony” to collapse. Speculation
around the cause of this disorder points to hive-diseases and pesticides harming the pollinators, though no
overall consensus has been reached. Twelve years later, some industries are observing recovery but the
American honey industry is still largely struggling. The U.S. used to locally produce over half the honey it
consumes per year. Now, honey mostly comes from overseas, with 350 of the 400 million pounds of honey
consumed every year originating from imports. This dataset provides insight into honey production supply and
demand in America by state from 1998 to 2012.

#### 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
A = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year']
sns.pairplot(a[A], hue="year")
plt.show()
```

# 
From the pairplot, we can see the following relationships:

There is a positive linear relationship between "numcol" and "total prod". As the number of colonies increases, the total production also tends to increase.
There is a positive linear relationship between "yield percol" and "total prod". As the yield per colony increases, the total production also tends to increase.
There is a positive linear relationship between "total prod" and "prodvalue". As the total production increases, the production value also tends to increase.
There is a positive linear relationship between "numcol" and "prodvalue". As the number of colonies increases, the production value also tends to increase.
In addition, we can see that the relationships between these variables remain relatively consistent across different years. This suggests that these relationships are relatively stable over time and may be useful for predicting future outcomes.

#### 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue'))

```python
corr = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
correlation=a[corr]
corr_matrix = correlation.corr()
sns.heatmap(corr_matrix, annot= True);
```

```python
From the heatmap, we can see the following correlation values:

There is a strong positive correlation (0.91) between "numcol" and "total prod".
There is a moderate positive correlation (0.62) between "yield percol" and "total prod".
There is a strong positive correlation (0.85) between "total prod" and "prodvalue".
There is a moderate negative correlation (-0.50) between "stocks" and "price per lb".
There is a moderate negative correlation (-0.61) between "price per lb" and "prodvalue".
In addition, we can see that there are no strong negative correlations between any of the variables.
This suggests that there are no variables that are strongly negatively related to each other.
Overall, the heatmap provides a comprehensive visualization of the correlation values between multiple columns.
```
